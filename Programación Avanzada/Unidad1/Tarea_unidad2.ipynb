```python
import math

# Definición de la clase Circulo
class Circulo:
    def __init__(self, radio):
        self.radio = radio
        self.perimetro = 2 * math.pi * radio
        self.area = math.pi * radio ** 2

# Pedir al usuario que ingrese el radio
radio_usuario = float(input("Ingrese el radio del círculo: "))

# Instancia de la clase Circulo con el radio proporcionado
circulo = Circulo(radio_usuario)

# Imprimir el perímetro y el área
print(f"Perímetro: {circulo.perimetro:.2f}")
print(f"Área: {circulo.area:.2f}")

```

```python
import math

class Punto3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def distancia(self, otro_punto):
        return math.sqrt((self.x - otro_punto.x)**2 + (self.y - otro_punto.y)**2 + (self.z - otro_punto.z)**2)

A = Punto3D(1, 2, 0)
B = Punto3D(3, 0, 0)
print(A.x, A.y, A.distancia(B))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
f_x = x**2 + 2*x + 3
g_x = -x**2 + 1

plt.plot(x, f_x, 'r--', label='f(x) = x^2 + 2x + 3')
plt.plot(x, g_x, 'g-', label='g(x) = -x^2 + 1')
plt.title('Gráfica de f(x) y g(x)')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Función para calcular los vértices de una estrella de 5 puntas
def estrella_5_puntas(radio=1, centro=(0, 0)):
    angulos = np.linspace(0, 2 * np.pi, 11)  # 11 puntos para hacer una estrella de 5 puntas
    angulos = angulos[:-1]  # Quitamos el último punto, porque coincide con el primero
    coordenadas = []

    for i, angulo in enumerate(angulos):
        if i % 2 == 0:
            r = radio  # Punto en el borde exterior
        else:
            r = radio / 2  # Punto en el interior
        x = centro[0] + r * np.cos(angulo)
        y = centro[1] + r * np.sin(angulo)
        coordenadas.append((x, y))

    return zip(*coordenadas)

# Coordenadas de la estrella
x, y = estrella_5_puntas()

# Graficar la estrella
plt.figure(figsize=(6,6))
plt.fill(x, y, color='yellow', edgecolor='orange', linewidth=3)
plt.title('Estrella de 5 puntas')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame con los datos proporcionados
data = {
    'Nombre': ['Ana', 'Luis', 'Rosa', 'Mario', 'Marisa'],
    'Edad': [23, 45, 35, 50, 28],
    'Altura': [1.60, 1.80, 1.55, 1.75, 1.68],
    'Colesterol': [200, 150, 180, 220, 190],
    'Sexo': ['F', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)


```

```python
# (a) Calcule el promedio de las edades
promedio_edades = df['Edad'].mean()
print(f"Promedio de las edades: {promedio_edades}")

```

```python
# (b) Muestre solo a las mujeres y calcule el promedio de su colesterol
mujeres = df[df['Sexo'] == 'F']
promedio_colesterol_mujeres = mujeres['Colesterol'].mean()
print(f"Promedio de colesterol en mujeres: {promedio_colesterol_mujeres}")

```

```python
# (c) Realice un gráfico de barras por persona y nivel de colesterol
df.plot(kind='bar', x='Nombre', y='Colesterol', color='blue', title='Colesterol por Persona')
plt.ylabel('Nivel de Colesterol')
plt.xlabel('Nombre')
plt.grid(True)
plt.show()

```

```python
# (d) Resalte la edad de Rosa y la altura de Marisa con color amarillo
def resaltar(df):
    colors = []
    for _, row in df.iterrows():
        if row['Nombre'] == 'Rosa':
            colors.append(['background-color: yellow' if col == 'Edad' else '' for col in df.columns])
        elif row['Nombre'] == 'Marisa':
            colors.append(['background-color: yellow' if col == 'Altura' else '' for col in df.columns])
        else:
            colors.append(['' for col in df.columns])
    return pd.DataFrame(colors, columns=df.columns)

df_style = df.style.apply(resaltar, axis=None)
df_style  # Mostrar el DataFrame con el estilo

```

```python
# (e) Ordene la tabla de mayor a menor de acuerdo al colesterol
df_ordenado = df.sort_values(by='Colesterol', ascending=False)
print("\nTabla ordenada de mayor a menor por colesterol:")
print(df_ordenado)

```
