# Gráficas de una Elipse y una Función en 3D




```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores de x y y para la elipse
t = np.linspace(0, 2 * np.pi, 100)
x = 3 * np.cos(t)
y = 5 * np.sin(t)

# Crear la figura y el gráfico
plt.figure()
plt.plot(x, y, color='green')  # Gráfica de la elipse en color verde


# título y etiquetas
plt.title(r'Gráfica de $\frac{x^2}{9} + \frac{y^2}{25} = 1$',fontsize = 15)
plt.xlabel('Eje x')
plt.ylabel('Eje y')

# Mostrar la gráfica
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

#  Datos para la gráfica
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x, y)
z = 9*x**2 - 5*y**2

# Crear la figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# superficie
ax.plot_surface(x, y, z, cmap='viridis')

# título y etiquetas de los ejes
ax.set_title('Gráfica 3D de $f(x, y) = 9x^2 - 5y^2$')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_zlabel('f(x, y)')

# Mostrar la gráfica
plt.show()
```
